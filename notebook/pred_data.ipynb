```python
import requests
dataaqi=requests.get('http://api.openweathermap.org/data/2.5/air_pollution?lat=18.9766&lon=72.8338&appid=74dcc4c2667120c8e4f27d2b7482d232').json()
```

```python
import datetime
pm2_5=dataaqi['list'][0]['components']['pm2_5']
pm10=dataaqi['list'][0]['components']['pm10']
time=dataaqi['list'][0]['dt']
time=datetime.datetime.fromtimestamp(time).strftime('%Y-%m-%d %H:%M:%S')
print(f"PM2.5: {pm2_5} µg/m³")
print(f"PM10: {pm10} µg/m³")
print(f"Data Timestamp: {time}")
```

```python
dataweather=requests.get('https://api.openweathermap.org/data/2.5/weather?lat=18.9766&lon=72.8338&appid=74dcc4c2667120c8e4f27d2b7482d232').json()
```

```python
temp=dataweather['main']['temp']-273
humidity=dataweather['main']['humidity']
windS=dataweather['wind']['speed']
windD=dataweather['wind']['deg']
print(f"Temperature: {temp:.2f} °C")
print(f"Humidity: {humidity} %")
print(f"Wind Speed: {windS} m/s")
print(f"Wind Direction: {windD} °")
```
